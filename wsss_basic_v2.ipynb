# Super basic wsss

Simply DinoV3 as encoder
Decoder as decoder

```python
class SingleDecoderWSS(nn.Module):
    def __init__(self):
        self.encoder = DINOv2()   # frozen
        self.decoder = SegDecoder(num_classes=20)  # no background yet

    def forward(self, x):
        tokens = self.encoder(x)
        seg_map = self.decoder(tokens)        # [B, 20, H, W]
        logits = F.adaptive_avg_pool2d(seg_map, 1).flatten(1)  # [B, 20]
        return seg_map, logits

# Loss: only BCE on logits
# At inference: argmax(seg_map) is your segmentation
```
