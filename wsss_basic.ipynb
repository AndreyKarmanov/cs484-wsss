```python
import torch.nn as nn

class CAMDecoder(nn.Module):
    def __init__(self, embed_dim=768, num_classes=20):
        super().__init__()
        self.conv = nn.Conv2d(embed_dim, num_classes, kernel_size=1)

    def forward(self, patch_tokens):
        B, N, C = patch_tokens.shape
        h = w = int(N ** 0.5)
        x = patch_tokens.permute(0, 2, 1).reshape(B, C, h, w)
        return self.conv(x)
```
